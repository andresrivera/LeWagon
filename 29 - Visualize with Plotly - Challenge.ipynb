c) Create a first graphical object of type Scatter which will be a single line between points A(2, 7) and B(3, 8). Store this object in a variable `first_line_object`.

```python
import plotly.graph_objects as go

# Define the coordinates of points A and B
x_coords = [2, 3]
y_coords = [7, 8]

# Create a Scatter plot with a line connecting points A and B
first_line_object = go.Scatter(x=x_coords, y=y_coords, mode='lines')

# Create a new figure called fig and add the first line object to it
fig = go.Figure(first_line_object)
```

d) Add this variable to your figure using the add_trace() method and then view the result with the show() method.

```python
import plotly.graph_objects as go

# Define the coordinates of points A and B
x_coords = [2, 3]
y_coords = [7, 8]

# Create a Scatter plot with a line connecting points A and B
first_line_object = go.Scatter(x=x_coords, y=y_coords, mode='lines')

# Create a new figure called fig
fig = go.Figure()

# Add the first_line_object to the figure
fig.add_trace(first_line_object)

# View the result
fig.show()
```

e) Graphical objects allow to realize very complex graphs by adding multiple graphical objects, but the writing is a bit heavy.

That's why Plotly Express library has been developed to save time and flexibility when manipulating DataFrame. It is always better to use Plotly Express first and if it is not enough to achieve what you want then use graphical objects.

Make the same graph as before but with the `line` function of Plotly Express.

```python
import plotly.express as px

# Define the coordinates of points A and B
x_coords = [2, 3]
y_coords = [7, 8]

# Create a DataFrame containing the coordinates of points A and B
df = pd.DataFrame({'x': x_coords, 'y': y_coords})

# Create a line plot using Plotly Express
fig = px.line(df, x='x', y='y')

# View the result
fig.show()
```

g) Using the `bar` function, represent as a bar the evolution of the Colombian life expectancy through time.

```python
import plotly.express as px

# Load the Gapminder dataset
df = px.data.gapminder()

# Filter the dataset for Colombia
colombia_data = df[df['country'] == 'Colombia']

# Create a bar plot for Colombian life expectancy over time
fig = px.bar(colombia_data, x='year', y='lifeExp', title='Evolution of Colombian Life Expectancy')
fig.show()
```

h) We want to make the same chart for the countries Colombia, Ecuador and Peru.

Look at the documentation to make 3 bar charts per year. We will use different colors for each country.

Hint : look at DataFrame's method isin() to filter on more than one value

```python
import plotly.express as px

# Load the Gapminder dataset
df = px.data.gapminder()

# Filter the dataset for Colombia, Ecuador, and Peru
countries = ['Colombia', 'Ecuador', 'Peru']
filtered_data = df[df['country'].isin(countries)]

# Create a bar chart for each country with different colors
fig = px.bar(filtered_data, x='year', y='lifeExp', color='country', title='Evolution of Life Expectancy (Colombia, Ecuador, Peru)')
fig.show()
```

i) We decide to change the representation by using line plots instead of bar plots.

Change the code to achieve this. Practical, isn't it?

```python
import plotly.express as px

# Load the Gapminder dataset
df = px.data.gapminder()

# Filter the dataset for Colombia, Ecuador, and Peru
countries = ['Colombia', 'Ecuador', 'Peru']
filtered_data = df[df['country'].isin(countries)]

# Create a line chart for each country with different colors
fig = px.line(filtered_data, x='year', y='lifeExp', color='country', title='Evolution of Life Expectancy (Colombia, Ecuador, Peru)')
fig.show()
```

Display for year 2007 only and for continent "Americas" only, the relation (find the right graph to visualize the correlations) between the variables `gfpPercap` (PIP per capita) and `lifeExp` with colors differentiated by country and a size of points proportional to the size of the population of the concerned country.

The argument `size_max` can be set to 60 to increase the size of the points while keeping the same proportions.

```python
import plotly.express as px

# Load the Gapminder dataset
df = px.data.gapminder()

# Filter the dataset for the year 2007 and the continent "Americas"
filtered_data = df[(df['year'] == 2007) & (df['continent'] == 'Americas')]

# Create a scatter plot
fig = px.scatter(filtered_data, 
                 x='gdpPercap', 
                 y='lifeExp', 
                 size='pop', 
                 color='country', 
                 title='Relationship between GDP per Capita and Life Expectancy (2007 - Americas)',
                 size_max=60)

# Show the plot
fig.show()
```
