2) How is data structured in this file ?

```python
import json

def print_catalogue_structure(data, indent=0):
    for key, value in data.items():
        if isinstance(value, dict):
            print('  ' * indent + f'Category: {key}')
            print_catalogue_structure(value, indent + 1)
        elif isinstance(value, list):
            print('  ' * indent + f'Subcategory: {key}')
            for product in value:
                print('  ' * (indent + 1) + f'Product: {product}')
        else:
            print('  ' * indent + f'Unknown data type for key: {key}')

def main():
    # Read the JSON file
    file_name = 'daco_fc_products_catalogue.json'
    with open(file_name, 'r') as file:
        data = json.load(file)

    # Print the catalogue structure
    print("Catalogue Structure:")
    print_catalogue_structure(data)

if __name__ == "__main__":
    main()
```

Main categories are keys, sub categories as sub keys and products are stored in lists.
